```python
#Para saber el cambio diario
cambiodiario = []
for i in range(7769) :
    buy = (data['adj_close'].iloc[i]/ data['open'].iloc[i])-1
    cambiodiario.append(buy)
```

```python
#Si el precio de apertura cambia un 10% se guarda para comprar 
compra = []
for i in range(7769):
    if cambiodiario[i]> 0.01:
        compra.append(i)
```

```python
# Elementos que cumplen con la condición
sicompra = pd.DataFrame(columns=('ticker', 'adj_close', 'open'))
for i in compra:
    sicompra.loc[i]=(data.iloc[i])
```
